```python
import pandas as pd
import numpy as np

DATA_PATH = "/kaggle/input/competitions/h-and-m-personalized-fashion-recommendations/"

customers = pd.read_csv(
    DATA_PATH + "customers.csv"
)

articles = pd.read_csv(
    DATA_PATH + "articles.csv"
)

transactions = pd.read_csv(
    DATA_PATH + "transactions_train.csv"
)
```

```python
print("Customers Missing Values")
print(customers.isnull().sum())

print("\nArticles Missing Values")
print(articles.isnull().sum())

print("\nTransactions Missing Values")
print(transactions.isnull().sum())
```

**Handle FN and Active**

These are mostly binary/flag columns.

```python
customers['FN'] = customers['FN'].fillna(0)

customers['Active'] = customers['Active'].fillna(0)
```

**Handle club_member_status**

Categorical → fill with "UNKNOWN"

```python
customers['club_member_status'] = (
    customers['club_member_status']
    .fillna("UNKNOWN")
)
```

**Handle fashion_news_frequency**

Categorical → fill with "NONE"

```python
customers['fashion_news_frequency'] = (
    customers['fashion_news_frequency']
    .fillna("NONE")
)
```

**Handle age**

Numerical → median is best.

```python
customers['age'] = (
    customers['age']
    .fillna(
        customers['age'].median()
    )
)
```

**Handle detail_desc**

Text description

```python
articles['detail_desc'] = (
    articles['detail_desc']
    .fillna("No description available")
)
```

```python
#Verify the fix
print("Customers Missing Values")
print(customers.isnull().sum())

print("\nArticles Missing Values")
print(articles.isnull().sum())

print("\nTransactions Missing Values")
print(transactions.isnull().sum())
```

**Duplicates handling**

```python
print("Customer duplicates:",
      customers.duplicated().sum())

print("Article duplicates:",
      articles.duplicated().sum())

print("Transaction duplicates:",
      transactions.duplicated().sum())
```

```python
transactions = transactions.drop_duplicates()
```
