```python
import pandas as pd
import pickle
import numpy as np
```

```python
with open(
    "/kaggle/input/datasets/selinparmar/item-and-user-score/item_store.pkl",
    "rb"
) as file:

    item_store = pickle.load(
        file
    )

with open(
    "/kaggle/input/datasets/selinparmar/item-and-user-score/user_store.pkl",
    "rb"
) as file:

    user_store = pickle.load(
        file
    )

print(
    "Files loaded!"
)
```

```python
print(
    "Total Items:",
    len(
        item_store
    )
)

print(
    "Total Users:",
    len(
        user_store
    )
)
```

```python
redis_users = {

    user[
        'customer_id'
    ]:

    user

    for user in
    user_store
}

redis_items = {

    item[
        'article_id'
    ]:

    item[
        'embedding'
    ]

    for item in
    item_store
}
```

```python
sample_user_id = (
    user_store[0][
        'customer_id'
    ]
)

redis_users[
    sample_user_id
]
```

```python
sample_article_id = (
    item_store[0][
        'article_id'
    ]
)

redis_items[
    sample_article_id
][:5]
```

```python
with open(
    "/kaggle/working/redis_users.pkl",
    "wb"
) as file:

    pickle.dump(
        redis_users,
        file
    )

with open(
    "/kaggle/working/redis_items.pkl",
    "wb"
) as file:

    pickle.dump(
        redis_items,
        file
    )

print(
    "Redis dictionaries saved!"
)
```
