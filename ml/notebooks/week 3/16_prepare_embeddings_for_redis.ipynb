```python
import pandas as pd
import pickle
import numpy as np
```

```python
training_df = pd.read_parquet(
    "/kaggle/input/datasets/selinparmar/train-ds/training_dataset.parquet"
)

training_df = (
    training_df
    .sample(
        20000,
        random_state=42
    )
)

print(
    training_df.shape
)
```

```python
item_embeddings_df = pd.read_pickle(
    "/kaggle/input/datasets/selinparmar/item-embeddings/item_embeddings.pkl"
)

item_embeddings_df.head()
```

```python
item_embeddings_df[
    'embedding'
] = (

    item_embeddings_df[
        'embedding'
    ]

    .apply(
        lambda x:
        x.tolist()
    )
)

print(
    "Converted embeddings!"
)
```

```python
item_store = []

for _, row in item_embeddings_df.iterrows():

    item_store.append({

        "article_id":
            str(
                row[
                    'article_id'
                ]
            ),

        "embedding":
            row[
                'embedding'
            ]
    })

print(
    len(
        item_store
    )
)
```

```python
user_profiles = (

    training_df[

        [
            'customer_id',
            'favorite_product_type',
            'season',
            'age',
            'purchase_frequency',
            'avg_spend',
            'repeat_purchase_ratio'
        ]

    ]

    .drop_duplicates(
        subset=
        'customer_id'
    )
)
```

```python
user_store = []

for _, row in user_profiles.iterrows():

    user_store.append({

        "customer_id":
            str(
                row[
                    'customer_id'
                ]
            ),

        "favorite_product_type":
            row[
                'favorite_product_type'
            ],

        "season":
            row[
                'season'
            ],

        "age":
            float(
                row[
                    'age'
                ]
            ),

        "purchase_frequency":
            float(
                row[
                    'purchase_frequency'
                ]
            ),

        "avg_spend":
            float(
                row[
                    'avg_spend'
                ]
            ),

        "repeat_purchase_ratio":
            float(
                row[
                    'repeat_purchase_ratio'
                ]
            )
    })

print(
    len(
        user_store
    )
)
```

```python
with open(
    "/kaggle/working/item_store.pkl",
    "wb"
) as file:

    pickle.dump(
        item_store,
        file
    )

with open(
    "/kaggle/working/user_store.pkl",
    "wb"
) as file:

    pickle.dump(
        user_store,
        file
    )

print(
    "Redis-ready files saved!"
)
```
