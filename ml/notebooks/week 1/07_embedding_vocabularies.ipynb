```python
import pandas as pd
import pickle
```

```python
training_df = pd.read_parquet(
    "/kaggle/input/datasets/selinparmar/training-dataset/training_dataset.parquet"
)

print(
    training_df.shape
)
```

```python
user_vocab = (
    training_df[
        'customer_id'
    ]
    .astype(str)
    .unique()
    .tolist()
)

print(
    len(user_vocab)
)
```

```python
item_vocab = (
    training_df[
        'article_id'
    ]
    .astype(str)
    .unique()
    .tolist()
)

print(
    len(item_vocab)
)
```

```python
product_type_vocab = (
    training_df[
        'product_type_name'
    ]
    .astype(str)
    .unique()
    .tolist()
)

print(
    product_type_vocab[:10]
)
```

```python
favorite_product_vocab = (
    training_df[
        'favorite_product_type'
    ]
    .astype(str)
    .unique()
    .tolist()
)

print(
    favorite_product_vocab[:10]
)
```

```python
season_vocab = (
    training_df[
        'season'
    ]
    .astype(str)
    .unique()
    .tolist()
)

print(
    season_vocab
)
```

```python
day_vocab = (
    training_df[
        'day_of_week'
    ]
    .astype(str)
    .unique()
    .tolist()
)

print(
    day_vocab
)
```

```python
feature_vocab = {

    'user_vocab':
        user_vocab,

    'item_vocab':
        item_vocab,

    'product_type_vocab':
        product_type_vocab,

    'favorite_product_vocab':
        favorite_product_vocab,

    'season_vocab':
        season_vocab,

    'day_vocab':
        day_vocab
}
```

```python
import pickle

with open(
    "/kaggle/working/feature_vocab.pkl",
    "wb"
) as file:

    pickle.dump(
        feature_vocab,
        file
    )

print("Vocabulary saved!")
```

```python
print(
    "Users:",
    len(user_vocab)
)

print(
    "Items:",
    len(item_vocab)
)

print(
    "Product Types:",
    len(product_type_vocab)
)

print(
    "Favorite Product Types:",
    len(favorite_product_vocab)
)

print(
    "Seasons:",
    season_vocab
)

print(
    "Days:",
    day_vocab
)
```
