```python
import pandas as pd
import numpy as np
```

```python
customer_behavior_features = pd.read_parquet(
    "/kaggle/input/datasets/selinparmar/dataset-features/customer_behavior_features.parquet"
)

article_features = pd.read_parquet(
    "/kaggle/input/datasets/selinparmar/dataset-features/article_features.parquet"
)

transactions = pd.read_parquet(
    "/kaggle/input/datasets/selinparmar/dataset-features/transactions_cleaned.parquet"
)

print("Datasets loaded!")
```

```python
#create time features
transactions['year'] = (
    transactions['t_dat']
    .dt.year
)

transactions['month'] = (
    transactions['t_dat']
    .dt.month
)

transactions['day_of_week'] = (
    transactions['t_dat']
    .dt.day_name()
)
```

```python
#create interaction label
transactions[
    'interaction_label'
] = 1
```

```python
transactions_sample = (
    transactions
    .sample(
        n=1000000,
        random_state=42
    )
)

print(
    transactions_sample.shape
)
```

```python
training_df = (
    transactions_sample
    .merge(
        customer_behavior_features,
        on='customer_id',
        how='left'
    )
)

training_df.head()
```

```python
training_df = (
    training_df
    .merge(
        article_features,
        on='article_id',
        how='left'
    )
)

training_df.head()
```

```python
training_df.columns
```

```python
training_df = training_df.rename(
    columns={
        'product_type_name_x':
            'favorite_product_type',

        'product_type_name_y':
            'product_type_name',

        'sales_channel_id_x':
            'sales_channel_id'
    }
)
```

```python
training_df = (
    training_df
    .drop(
        columns=['sales_channel_id_y'],
        errors='ignore'
    )
)
```

```python
training_df.columns
```

```python
selected_columns = [

    'customer_id',
    'article_id',

    'age',

    'purchase_frequency',
    'days_since_last_purchase',
    'avg_spend',
    'avg_days_between_purchases',
    'repeat_purchase_ratio',

    'favorite_product_type',
    'product_type_name',

    'season',
    'sales_channel_id',

    'purchase_count',

    'month',
    'day_of_week',

    'interaction_label'
]

training_df = (
    training_df[
        selected_columns
    ]
)

training_df.head()
```

```python
text_cols = [
    'favorite_product_type',
    'product_type_name',
    'season'
]

training_df[
    text_cols
] = (
    training_df[
        text_cols
    ]
    .fillna("Unknown")
)

numeric_cols = [
    'age',
    'purchase_frequency',
    'days_since_last_purchase',
    'avg_spend',
    'avg_days_between_purchases',
    'repeat_purchase_ratio',
    'purchase_count'
]

training_df[
    numeric_cols
] = (
    training_df[
        numeric_cols
    ]
    .fillna(0)
)
```

```python
training_df.info()
```

```python
training_df.to_parquet(
    "/kaggle/working/training_dataset.parquet",
    index=False
)

print("Training dataset saved!")
```
